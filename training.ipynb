```python
import kagglehub
path = kagglehub.dataset_download("yehyichang/tea-leaf-diseases")
print("Returned path:", path)
```

```python
import os, glob

dataset_root = "/root/.cache/kagglehub/datasets/yehyichang/tea-leaf-diseases/versions/1/tealeafdiseases.v3i.yolov5pytorch"


print("Top-level dirs:", os.listdir(dataset_root))


images = glob.glob(os.path.join(dataset_root, "**", "*.jpg"), recursive=True)
txts = glob.glob(os.path.join(dataset_root, "**", "*.txt"), recursive=True)

print("Total images:", len(images))
print("Total YOLO txt labels:", len(txts))


print("Sample images:", images[:5])
print("Sample txts:", txts[:5])
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")


data_yaml = "/root/.cache/kagglehub/datasets/yehyichang/tea-leaf-diseases/versions/1/tealeafdiseases.v3i.yolov5pytorch/data.yaml"


results = model.train(
    data=data_yaml,
    epochs=50,    
    imgsz=640,    
    batch=16 )
```
